```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
df_pt_1 = pd.read_csv('C:/Users/vasco/Ironhack/labs/Week5/Second Project/second_project/data/raw/df_final_web_data_pt_1.txt')
df_pt_2 = pd.read_csv('C:/Users/vasco/Ironhack/labs/Week5/Second Project/second_project/data/raw/df_final_web_data_pt_2.txt')
```

```python
df_combined = pd.concat([df_pt_1, df_pt_2], ignore_index=True)
```

```python
# Count the duplicates
duplicate_count = df_combined.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
```

```python
# Drop duplicates
df_combined = df_combined.drop_duplicates()

# Resetindex
df_combined.reset_index(drop=True, inplace=True)
```

```python
# Drop duplicate rows
df_combined = df_combined.drop_duplicates()

# Convert 'date_time' column to datetime format
df_combined.loc[:, 'date_time'] = pd.to_datetime(df_combined['date_time'])

# Normalize the casing in 'process_step'
df_combined.loc[:, 'process_step'] = df_combined['process_step'].str.lower()

# Unique values in 'process_step'
print(df_combined['process_step'].value_counts())

# Check numerical data
print(df_combined.describe())
```

```python
#Dataset sort by 'client_id', 'visit_id', 'process_step', and 'date_time'
df_combined.sort_values(by=['client_id', 'visit_id', 'process_step', 'date_time'], ascending=True, inplace=True)
```

```python
df_combined
```

```python
#Filter data to consider only the last start time per client
# Group by 'client_id' and 'visit_id', and consider the last occurrence of 'start' (first row in sorted data)
df_start = df_combined[df_combined['process_step'] == 'start'].groupby('client_id').last().reset_index()
```

```python
df_start
```

```python
#Filter data to consider only the last 'confirm' (end step) per client
df_confirm = df_combined[df_combined['process_step'] == 'confirm'].groupby('client_id').last().reset_index()
```

```python
df_confirm
```

```python
#Join the start and confirm data
df_final = pd.merge(df_start, df_confirm, on='client_id', suffixes=('_start', '_confirm'))
```

```python
df_final
```

```python
#Calculate completion rates (number of clients who completed each step)
clients_started = df_combined[df_combined['process_step'] == 'start']['client_id'].nunique()
completion_rates = (
    df_combined.groupby('process_step')['client_id'].nunique() / clients_started
).reset_index(name='completion_rate')
```

```python
completion_rates
```

```python
# Sort Data
df_combined = df_combined.sort_values(by=['client_id', 'visit_id', 'date_time'])

# Time difference to next step (same client)
df_combined['time_diff'] = df_combined.groupby(['client_id', 'visit_id'])['date_time'].diff()

# Average time spent on each step
completion_times = df_combined.groupby('process_step')['time_diff'].mean()
```

```python
completion_times
```

```python
step_order = {'start': 0, 'step_1': 1, 'step_2': 2, 'step_3': 3, 'confirm': 4}
df_combined['step_numeric'] = df_combined['process_step'].map(step_order)
# Ensure the data is sorted by client_id and date_time
df_combined = df_combined.sort_values(by=['client_id', 'date_time'])
# Check if the next step is less than the current step
df_combined['error'] = df_combined.groupby('client_id')['step_numeric'].shift(-1) < df_combined['step_numeric']
error_rate = df_combined['error'].mean()
```
